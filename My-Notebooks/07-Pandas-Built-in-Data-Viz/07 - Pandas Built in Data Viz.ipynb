# Pandas Built-in Data Visualization

```python
#built off matplotlib
#allows you to create plots right on pandas dataframes
```

```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
import seaborn as sns
```

```python
df1 = pd.read_csv('df1', index_col = 0)
df1.head()
```

```python
df2 = pd.read_csv('df2')
df2.head()
```

```python
#get a histogram for the a column in df1
df1['A'].hist()
```

```python
#can add matplotlib and seaborn arguments into the graph
df1['A'].hist(bins = 30)
```

```python
df1['A'].plot(kind = 'hist', bins = 30)
#same as df1['A'].hist(bins = 30)
```

```python
df1['A'].plot.hist()
```

#### Area plot

```python
df2.plot.area(alpha = 0.4)
```

#### bar plot

```python
df2.plot.bar()
#index value is a category that gets split up for each x value
```

```python
df2.plot.bar(stacked = True)
```

#### lineplot

```python
df1.plot.line(y = 'B')
```

```python
df1.plot.line(y = 'B', figsize = (12,3), linewidth = 0.1)
```

#### scatterplots

```python
df1.plot.scatter(x = 'A', y = 'B')
```

```python
#can set the color based off another column
df1.plot.scatter(x = 'A', y = 'B', c = 'C', cmap = 'coolwarm')
#3 levels of information
```

```python
#can set the size of marker based on another column
df1.plot.scatter(x = 'A', y = 'B', s = df1['C']*100)
```

#### boxplots

```python
df2.plot.box()
```

#### hexbinplot

```python
df = pd.DataFrame(np.random.randn(1000,2), columns = ['a','b'])
df.head()
```

```python
df.plot.hexbin(x = 'a', y = 'b', gridsize = 25, cmap = 'magma')
#colored by the number of points within each bin
```

#### kde plots

```python
df2['a'].plot.kde()
```

```python
df2['a'].plot.density()
```

```python
df2.plot.kde()
```
