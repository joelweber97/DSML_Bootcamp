```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

#### Functional method

```python
x = np.linspace(0,5,num = 11)
y = x**2
```

```python
#the most basic plot
plt.plot(x, y)
```

```python
#adding an x and y label and title
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
#multiplots on the same canvas
plt.subplot(1, 2, 1)  #plt.subplot(#rows, #columns, plot #) -> 1 row, 2 columns, plot 1
plt.plot(x,y,'r')
plt.subplot(1,2,2) #plt.subplot(#rows, #columns, plot #) -> 1 row, 2 columns, plot 2
plt.plot(y,x,'b')
```

#### Object oriented api method

```python
#main idea is to create figure objects and then call methods on those
```

```python
fig = plt.figure() #returns a figure object <Figure size 432x288 with 0 Axes>
#manually create a set of axes for fig
axes = fig.add_axes([0.1,0.1,0.8,0.8])  #[left, bottom, width, height]
#plot the data on the fig
axes.plot(x,y)
#add x label, y label, and title
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
#2 sets of figures on 1 canvas using the oo approach
fig = plt.figure()  #create figure object

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.25,0.5,0.4,0.3])
```

```python
fig = plt.figure()  #create figure object

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([.1,.5,.9,.5])  #starts 10% to the left, 50% from the bottom, 90% width, 50%height
```

```python
fig = plt.figure()  #create figure object

axes1 = fig.add_axes([.1,.1,.8,.8])
axes2 = fig.add_axes([.55,.1,.35,.6])
```

```python
#plotting on these axes
fig = plt.figure()  #create figure object

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) 
axes2 = fig.add_axes([0.25,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title('Bigger Plot')
axes2.set_title('Smaller Plot')
```

#### subplots using oop method

```python
plt.subplots(nrows = 1, ncols = 2)  #returns a tuple with (the figure object, [array of axes subplots])
```

```python
#we can use tuple unpacking to set fig to the figure object and axes to the 2 axes subplots 
fig, axes = plt.subplots(nrows=1,ncols=2) #returns 2 empty subplots
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2) #subplots basically calls fig.add_axes automatically for you
axes  #returns an array of AxesSubplot objects for each subplot and can iterate over them
for current in axes:
    current.plot(x,y)
```

```python
#since we can iterate through axes we can also index it as well.
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
plt.tight_layout()  #prevents overlapping text when dealing with subplots
```

#### Figure size and dpi

```python
fig = plt.figure(figsize = (3,2), dpi = 100)  #figsize = (width,height)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
#figsize is the size for all plots to fit into, not the size for each plot
fig,axes = plt.subplots(nrows = 2, ncols = 1, figsize = (8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

#### Saving a figure

```python
#png, jpeg, svg, pdf, etc
```

```python
#code from above
fig,axes = plt.subplots(nrows = 2, ncols = 1, figsize = (8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

#### Legends

```python
#plotting 2 lines on the same figure
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2)
ax.plot(x,x**3)
```

```python
#now lets add a legend to make it easy to see what line is what
#also need to include a label for each ax.plot so it knows what to call it in the legend
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'x**2')
ax.plot(x,x**3, label = 'x**3')
ax.legend()
```

```python
#can define the location of the legend using the loc parameter in ax.legend()
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'x**2')
ax.plot(x,x**3, label = 'x**3')
ax.legend(loc = 0)  #'best/0','upper right/1', 'center/10', etc
```

```python
#can define the location of the legend using the loc parameter in ax.legend()
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'x**2')
ax.plot(x,x**3, label = 'x**3')
ax.legend(loc = 'lower center')  #'best/0','upper right/1', 'center/10', etc
```

```python
#can define the location of the legend using the loc parameter in ax.legend()
#if the set locations aren't ideal we can pass a tuple in 
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'x**2')
ax.plot(x,x**3, label = 'x**3')
ax.legend(loc = (0.1,0.1))  #passing a tuple in to define our own location
```

#### Plot Appearance

```python
#setting colors
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = '#FF8C00')  #common color strings ('orange'), rgb hex codes ('#FF8C00'), etc
```

```python
#setting line width
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = '#FF8C00', linewidth = 3) #line width 3 times the default thickness
```

```python
#alpha argument changes line transparency
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = '#FF8C00', linewidth = 3, alpha = 0.5) #50% transparency
```

```python
#linestyle argument
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, linestyle = ':', linewidth = 3) 
#'-' solid line
#'--' dashed line
#'-.' dash dotted line
#':' dotted line
```

```python
#markers
#lets mark the x points on the plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, linestyle = ':', linewidth = 3, marker = '^') 
#possible markers are 'o', '*','.',',','^','x',etc
```

```python
#markersize and markerfacecolor, markeredgewidth, and markeredgecolor are also an argument
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, linestyle = '-', color = 'red', linewidth = 2, marker = '*', markersize = 20,
       markerfacecolor = 'yellow',markeredgewidth = 3, markeredgecolor = 'green') 
```

#### Axis appearance and plot range

```python
#y limit and x limit
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'purple')
ax.set_xlim([0,1])  #pass in a list of [lowbound,highbound]
ax.set_ylim([0,2]) #pass in a list of [lowbound,highbound]
```
