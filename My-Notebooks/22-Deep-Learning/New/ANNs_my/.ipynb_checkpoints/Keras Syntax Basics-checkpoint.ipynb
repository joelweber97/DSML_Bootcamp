## Part 1 - Preparing the Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/fake_reg.csv')
```

```python
df.head()
```

```python
#we will attempt to predict the price based of feature1 and feature2
```

```python
sns.pairplot(df)
#appears that feature 2 is highly correlated with price
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['feature1', 'feature2']].values  #need to pass in a numpy array so we add the .values to get an array back
y = df['price'].values
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
#once we have the split we will normalize/scale our data.
#this is necessary to prevent a column of large values from over affecting weights
from sklearn.preprocessing import MinMaxScaler
```

```python
help(MinMaxScaler)
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)  #only fit the scaler to the train set because we don't want our knowledge of the labels
#to have any influence on the model
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_train.max(), X_train.min()
```

## Part 2 - Creating and Training the Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#basically we set up a sequential model and then add layers to it
help(Sequential)
```

```python
'''model = Sequential()
model.add(Dense(4, activation = 'relu'))   
#units are the number of neurons while activation is the activation function
model.add(Dense(2, activation = 'relu'))   
model.add(Dense(1))
#this example model has 3 layers with 4, 2, and 1 neurons'''
```

```python
#back to our data
model = Sequential()
model.add(Dense(4, activation = 'relu'))
model.add(Dense(4, activation = 'relu'))
model.add(Dense(4, activation = 'relu'))
model.add(Dense(1))

model.compile(optimizer = 'rmsprop', loss = 'mse')
#for multi-class classification problems
#model.comple(optimizer = '...', loss = 'categorical_crossentroy', metrics = ['accuracy'])

#for binary classification use
#model.compile(optimizer = '...', loss = 'binary_crossentropy', metrics = ['accuracy'])

#for a mean squared error regression problem use
#model.compile(optimizer = '...', loss = 'mse')
```

```python
#now the model has been created we'll fit the model to the training data.

model.fit(x = X_train, y = y_train, epochs = 250, verbose = 1)
```

```python
#we can see the loss values decrease from 250,000 all the way down to 24
```

```python
#we can also visualize the loss values over the epochs
loss = pd.DataFrame(model.history.history)
loss.head()
```

```python
loss.plot()
```

```python
#we can see that at a certain number of epochs loss is not really decreasing all that much
```

## Part 3 - Model Evaluation

```python
#now we want to see how well the model performs with data it's never seen before
```

#### First Method of Evaluation

```python
model.evaluate(x = X_test, y = y_test)
#this returns the model's loss. In this case it returns the mean squared error of the test dataset.
```

```python
model.evaluate(x = X_train, y = y_train)
```

#### Second method of evaluation

```python
test_predictions = model.predict(X_test)
test_predictions
#for each value in the test dataset it runs it through the model to get a predicted value
```

```python
test_predictions = pd.Series(test_predictions.reshape(300,))
test_predictions
#we convert those predictions into a pandas series
```

```python
pred_df = pd.DataFrame(y_test, columns = ['Test True Y'])
pred_df
#We also create a dataframe from the real values of the test dataset
```

```python
pred_df = pd.concat([pred_df, test_predictions], axis = 1)
pred_df.columns = ['Test True Y', 'Model Predictions']
pred_df
#we join the true y value dataframe with the prediction series from above
#we can see the true value of the data compared to the prediction based on the model
```

```python
pred_df.info()
```

```python
#plot those true y values agains the predicted y values
sns.scatterplot(x = 'Test True Y', y = 'Model Predictions', data = pred_df)
#this shows the predicted value is actually pretty close to the real y value, meaning the model is performing very well
```

#### Third method of evaluation

```python
#we can also apply the regression evaluations (RMSE, MSE, MAE) on this scatterplot as well 
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(pred_df['Test True Y'], pred_df['Model Predictions'])
#means that on average the prediction is about 4 dollars off (really good in this instance)
```

```python
mean_squared_error(pred_df['Test True Y'], pred_df['Model Predictions'])
#MSE is pretty good too
```

```python
np.sqrt(mean_squared_error(pred_df['Test True Y'], pred_df['Model Predictions']))
#RMSE of 5.00 is also very good in this case
```

## Predicting on Brand New Data

```python
new_gem = [[998, 1000]]
```

```python
#since the model is based on the oringially scalde data, we have to take this new gem and scale it based on the original scaling
new_gem = scaler.transform(new_gem)
new_gem
```

```python
model.predict(new_gem)
#this newly found gem will sell for 419 dollars
```

```python
#the way you evaluate your test set is the same that you would use on a brand new data point
```
