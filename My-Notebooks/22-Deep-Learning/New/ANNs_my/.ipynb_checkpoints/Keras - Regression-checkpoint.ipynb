#### data preprocessing and creating a model

```python
df.head()
```

```python
#now the exploratory data analysis has been conducted we can split our features
X = df.drop('price', axis = 1).values
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
#train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)
```

```python
#now we'll perform the scaling on the training dataset 
#we don't want to scale before splitting to prevent leakage where test set affects the training model at all
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
#now we'll create the model
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#usually want to set up the model so the number of neurons is equal to the number of features in the df
X_train.shape
#we'll want 19 neurons to start with
```

```python
model = Sequential()

model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))

model.add(Dense(1))

model.compile(optimizer = 'adam', loss = 'mse')
```

```python
#train the model
model.fit(x = X_train, y = y_train, validation_data = (X_test, y_test), batch_size = 128 , epochs = 400)
```

setting validation data will record the loss at the end of each epoch so we can track how the model is doing on the training data AND on the test data (the test data will not influence weights and biases of the model)

Pass in batch size as powers of 2. The smaller the batch size the longer training is going to take but less likely you are to overfit to oyour data since you aren't passing in the entire dataset at once.

We don't have any stopping mechanisms yet (we will later in the course)

#### Model evaluation

```python
model.history.history
#returns a dictionary of the loss value of the training data and the loss value of the validation data
```

```python
loss = pd.DataFrame(model.history.history)
loss.head()
#turn that loss data into a df
```

```python
loss.plot()
#plot the losses against each other
```

Decrease in training and validation loss is exactly what we want to see.

No increase in validation loss really means we could continue training because the model isn't overfitting

As soon as the validation loss starts to increase we know that the model is being overfit and we need to stop it.

```python
#evaluation of our test data
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
```

```python
#create predictions from the test data
predictions = model.predict(X_test)
predictions
```

```python
#compare the predictions with the true label values
print(mean_squared_error(y_test, predictions))
print(np.sqrt(mean_squared_error(y_test, predictions)))
print(mean_absolute_error(y_test, predictions))
```

```python
df['price'].describe()
#since the mean price is 540k and our model is off by 106k on average
#so its not great because we're off by around 20%
```

```python
explained_variance_score(y_test, predictions)
#tells you how much variance is actually being explained by your model
#79% isnt great but not terrible either.
#one option is to keep training the model since we haven't overfit it yet.
```

```python
#can also plot our predictions against a perfict fit
plt.scatter(y_test, predictions)
plt.plot(y_test, y_test, 'r')
#not a great prediction but you can see that the really expensive outlier prices are throwing off the
#accuracy results. Pretty 
```

```python
#may be worth it to retrain our model witout those price outliers in it to see if that imporoves anything
```
