This lecture will show how to perform a classification task with tf

We will also focus on how to identify and deal with overfitting through Early Stopping Callbacks and Dropout Layers

Keras can automatically stop training based on a loss condition on the validation data passed during the model.fit() call.

Dropout layers can be added to "turn off" neurons during training to prevent overfitting
Each Dropout Layer will "drop" a user-defined percentage of neuron units in the previous layers every batch

#### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('benign_0__mal_1', axis = 1).values
y = df['benign_0__mal_1'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 101)
```

#### Scaling the Data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
#want to fit and transform the X_train data
```

```python
X_test = scaler.transform(X_test) 
#only want to transform the X_test data
```

#### Creating the Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout, Dense
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(30, activation = 'relu'))
model.add(Dense(15, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
#binary classification so we use the logiistic act. func.
model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
#binary classification uses binary_crossentropy loss
```

#### Fit the model to the data

```python
model.fit(x = X_train, y = y_train, epochs = 600, validation_data = (X_test, y_test))
#we're adding too many epochs to this model so we can see the validatation loss start to increase
#back up which signifies overfitting the model to the training data.
#only took 27 seconds on cpu while gpu took 51 seconds (600 epochs)
#only took 4 min 32 sec on cpu while gpu took 8 min 28 sec (6000 epochs)
```

#### Plotting the loss and validation loss

```python
losses = pd.DataFrame(model.history.history)
losses.head()
```

```python
losses.plot()
```

we can see the validation loss start to creep up indicating overfitting of the model to the training data.

We need to implement early stopping using TF callbacks feature that looks at increases in validation loss and stops the model early

#### Early stopping using TF Callbacks to prevent overfitting

```python
model = Sequential()

model.add(Dense(30, activation = 'relu'))
model.add(Dense(15, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
#binary classification so we use the logiistic act. func.
model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
#binary classification uses binary_crossentropy loss
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
help(EarlyStopping)
```

```python
early_stop = EarlyStopping(monitor = 'val_loss', mode = 'min', patience =25)
#want to minimize validation loss and we'll wait 25 epochs after the minimum value just in case of noise
```

#### Adding Dropout Layers to prevent overfitting

Dropout layers will turn off a user-defined percentage of neurons randomly.

The rate param is the Float between 0 and 1. Fraction of the input units to drop.  (rate = 0 means 0 percent of the neurons are being turned off, rate = 1 means 100 percent of neurons are being turned off randomly). Most common values range between 0.2 and 0.5.

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()

model.add(Dense(30, activation = 'relu'))
model.add(Dropout(rate = 0.5))
model.add(Dense(15, activation = 'relu'))
model.add(Dropout(rate = 0.5))
model.add(Dense(1, activation = 'sigmoid'))
#binary classification so we use the logiistic act. func.
model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
#binary classification uses binary_crossentropy loss
```

```python
early_stop = EarlyStopping(monitor = 'val_loss', mode = 'min', patience =25)
#want to minimize validation loss and we'll wait 25 epochs after the minimum value just in case of noise
```

```python
model.fit(x = X_train, y = y_train, epochs = 600, validation_data = (X_test, y_test), callbacks = [early_stop])
#callbacks passed in as a list
```

Now the model ran on 142 epochs before early stop kicked in. Running longer is a good thing because it means the model is still learning on the later epochs.

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

Exactly the kind of behavior we want to see.

#### Full evaluation on the classes

```python
predictions = model.predict_classes(X_test)
predictions
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```
